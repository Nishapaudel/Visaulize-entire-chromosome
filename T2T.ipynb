```python
import pandas as pd
import seaborn as sns
import py2bit
f_out = open('T2T.txt', 'w')

bd =py2bit.open('/home/ccmb/Downloads/T2T.2bit', True)
chroms = bd.chroms()
print(chroms)

print('CG','GC','NN' , sep = '\t', file = f_out)

start = 1976
end = start + 45090680 # this is the length of entire chromosome
for i in range(start, end , 10000 ):
    win_start = i 
    win_end = win_start + 1000
    perseq = bd.sequence('chr20', win_start, win_end)
    #print(perseq)
    A = int(perseq.upper().count('A'))
    T = int(perseq.upper().count('T'))
    C = int(perseq.upper().count('C'))
    G = int(perseq.upper().count('G'))
    CG = int(perseq.upper().count('CG'))
    GC = int(perseq.upper().count('GC'))
    NNN = int(perseq.upper().count('NNNNNNNNNN'))#25 Ns
    print(CG, GC, NNN, sep = '\t')
    print(CG ,GC,NNN, sep = '\t', file = f_out)
```
